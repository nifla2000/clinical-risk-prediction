Target Variable: The target variable is readmitted, which represents whether a patient was readmitted to the hospital after discharge. This project formulates the problem as a binary classification task, where patients readmitted within 30 days are considered high-risk cases

The dataset is imbalanced, with the majority of encounters resulting in no readmission. Readmissions within 30 days (<30) form the minority class and represent the clinically significant high-risk group. Class imbalance is important because accuracy alone may be misleading, requiring metrics such as recall and ROC-AUC.

```python
(df == '?').sum().sort_values(ascending=False).head(10)
```

```python
df = df.replace('?', np.nan)
```

```python
df.isna().sum().sort_values(ascending=False).head(10)
```

```python
df = df.drop(columns='weight')
```

glucose serum, A1 C results are missing because they are not measured as they were not required for those patients.

```python
df['max_glu_serum']= df['max_glu_serum'].fillna('Not_measured')
```

```python
df['A1Cresult']= df['A1Cresult'].fillna('Not_measured')
```

```python
df[['max_glu_serum', 'A1Cresult']].isna().sum()
```

```python
df['readmitted_binary'] = df['readmitted'].apply(
    lambda x: 1 if x == '<30' else 0
)
```

```python
df[['readmitted', 'readmitted_binary']].value_counts()
```

```python
df = df.drop(columns=['readmitted'])
```

```python
df['readmitted_binary'].value_counts()
```

```python
df['readmitted_binary'].value_counts(normalize=True)
```

The original multi-class target variable readmitted was transformed into a binary classification target.
Readmissions within 30 days were labeled as high-risk cases (1), while all other outcomes were labeled as low-risk (0).
This formulation reflects real-world clinical priorities and enables simpler and more interpretable models.
